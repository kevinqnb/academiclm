```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fname = '../data/pond_results.npz'
data = np.load(fname, allow_pickle=True)
responses = data['responses']
parametric_scores = data['parametric_scores']
context_scores = data['context_scores']
```

```python
responses
```

```python
parametric_score_min = 0
parametric_score_max = max([np.max(p) for p in parametric_scores])
context_score_min = 0
context_score_max = max([np.max(c) for c in context_scores])
```

```python
query_id = 2
print(f"Query {query_id+1}:")
print(responses[query_id])
```

```python
fig,ax = plt.subplots(figsize=(6,6))
pos = ax.matshow(context_scores[query_id], cmap='BuPu', vmin = context_score_min, vmax = context_score_max)
ax.set_ylabel('Layers')
ax.set_xlabel('Heads')
ax.set_title(f'Context Scores')
fig.colorbar(pos, ax=ax)
```

```python
fig,ax = plt.subplots(figsize=(6,4))
ax.bar(
    range(len(parametric_scores[query_id])), parametric_scores[query_id], alpha = 0.5
)
ax.set_ylim(parametric_score_min, parametric_score_max)
ax.set_ylabel('Parametric Score')
ax.set_xlabel('Layers')
```
